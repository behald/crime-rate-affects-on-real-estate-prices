```python
Q1 = df2['Population'].quantile(0.25)
Q3 = df2['Population'].quantile(0.75)

# Calculate the interquartile range (IQR)
IQR = Q3 - Q1

# Define the lower and upper bounds to identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers using the bounds
outliers = pd.DataFrame(df2[(df2['Population'] >= lower_bound) & (df2['Population'] <= upper_bound)])


# Remove outliers from the DataFrame
cleaned_df = outliers
    
    
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = cleaned_df['City, State']
values1 = cleaned_df['Homicides']
values2 = cleaned_df['Rapes']
values3 = cleaned_df['Assaults']
values4 = cleaned_df['Robberies']


# Creating positions for bars
bar_positions = np.arange(len(categories))
bar_width = 0.2  # Width of each bar

# Plotting the bars for each dataset
plt.figure(figsize=(10, 6))  # Optional: Adjusting figure size

plt.bar(bar_positions - 1.5 * bar_width, values1, bar_width, label='Value 1')
plt.bar(bar_positions - 0.5 * bar_width, values2, bar_width, label='Value 2')
plt.bar(bar_positions + 0.5 * bar_width, values3, bar_width, label='Value 3')
plt.bar(bar_positions + 1.5 * bar_width, values4, bar_width, label='Value 4')


# Adding x-axis labels
plt.xticks(bar_positions, categories)

# Adding title and labels
plt.title('Bar Graph with Four Columns')
plt.xlabel('Categories')
plt.ylabel('Values')

# Adding legend
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = cleaned_df['City, State']
values1 = cleaned_df['Homicides']
values2 = cleaned_df['Rapes']
values3 = cleaned_df['Assaults']
values4 = cleaned_df['Robberies']
values5 = cleaned_df['Violent Crimes']


# Creating positions for bars
bar_positions = np.arange(len(categories))

# Plotting the stacked bar chart
plt.figure(figsize=(15, 50))  # Optional: Adjusting figure size
plt.bar(bar_positions, values1, color='skyblue', label='Value 1')
plt.bar(bar_positions, values2, bottom=values1, color='orange', label='Value 2')
plt.bar(bar_positions, values3, bottom=np.array(values1) + np.array(values2), color='green', label='Value 3')
plt.bar(bar_positions, values4, bottom=np.array(values1) + np.array(values2) + np.array(values3), color='red', label='Value 4')
plt.bar(bar_positions, values5, bottom=np.array(values1) + np.array(values2) + np.array(values3) + np.array(values4), color='purple', label='Value 5')

# Adding x-axis labels
plt.xticks(bar_positions, categories)

# Adding title and labels
plt.title('Stacked Bar Chart with 5 Columns')
plt.xlabel('Categories')
plt.ylabel('Values')

# Adding legend
plt.legend()

# Show plot
plt.show()
```
